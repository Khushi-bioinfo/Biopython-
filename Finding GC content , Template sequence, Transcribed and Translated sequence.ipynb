```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction

fasta_file = "AJ009655.1.fasta"

for record in SeqIO.parse(fasta_file, "fasta"):
    print("Sequence ID:", record.id)
    print("Sequence:", record.seq)
    gc_percent = gc_fraction(record.seq) * 100
    print("GC% =", round(gc_percent, 2))
    print("---")
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
record = next(SeqIO.parse("AJ009655.1.fasta", "fasta"))
coding_dna = record.seq

template_dna = coding_dna.reverse_complement()

print("Template DNA:", template_dna)
```

```python
from Bio import SeqIO
record= next(SeqIO.parse ("AJ009655.1.fasta", "fasta"))
coding_dna = record.seq
messenger_rna= coding_dna.transcribe()
print("Messenger RNA:",messenger_rna)
```
